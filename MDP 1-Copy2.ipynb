```python
from hiive.mdptoolbox.mdp import ValueIteration, PolicyIteration, QLearning,PolicyIterationModified, ValueIterationGS
from hiive.mdptoolbox.example import forest
from hiive.mdptoolbox.openai import OpenAI_MDPToolbox
import numpy as np
import matplotlib.pyplot as plt
import time
import gym
import re
np.random.seed(42)
```

```python
x = OpenAI_MDPToolbox("FrozenLake8x8-v0")
value_f = []
policy = []
iters = []
time_array = []
gammaList = np.arange(0.05,1.0,.05)

epsilonList = np.arange(0.05,1.0,.05)


for i in gammaList:
    pi = PolicyIteration(x.P, x.R, i)
    pi.run()
    value_f.append(np.mean(pi.V))
    policy.append(pi.policy)
    iters.append(pi.iter)
    time_array.append(pi.time)

plt.plot(gammaList, time_array)
plt.xlabel('Gamma (Discount)')
plt.ylabel('Time (s)')
plt.grid()
plt.show()


plt.plot(gammaList,value_f)
plt.xlabel('Gamma (Discount)')
plt.ylabel('Average Rewards')
plt.grid()
plt.show()

plt.plot(gammaList,iters)
plt.xlabel('Gamma (Discount)')
plt.ylabel('Iterations')
plt.grid()
plt.show()

value_f = []
policy = []
iters = []
time_array = []

for i in epsilonList:
    pi = PolicyIterationModified(x.P, x.R, 0.95, epsilon=i)
    pi.run()
    value_f.append(np.mean(pi.V))
    policy.append(pi.policy)
    iters.append(pi.iter)
    time_array.append(pi.time)

plt.plot(epsilonList, time_array)
plt.xlabel('Epsilon')
plt.ylabel('Time (s)')
plt.grid()
plt.show()


plt.plot(epsilonList,value_f)
plt.xlabel('Epsilon')
plt.ylabel('Average Rewards')
plt.grid()
plt.show()

plt.plot(epsilonList,iters)
plt.xlabel('Epsilon')
plt.ylabel('Iterations')
plt.grid()
plt.show()
```
